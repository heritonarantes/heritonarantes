### *O que é o Covid ?*

A Covid-19 é uma infecção respiratória aguda causada pelo coronavírus SARS-CoV-2, potencialmente grave, de elevada transmissibilidade e de distribuição global.

O SARS-CoV-2 é um betacoronavírus descoberto em amostras de lavado broncoalveolar obtidas de pacientes com pneumonia de causa desconhecida na cidade de Wuhan, província de Hubei, China, em dezembro de 2019. Pertence ao subgênero Sarbecovírus da família Coronaviridae e é o sétimo coronavírus conhecido a infectar seres humanos.

Os coronavírus são uma grande família de vírus comuns em muitas espécies diferentes de animais, incluindo o homem, camelos, gado, gatos e morcegos. Raramente os coronavírus de animais podem infectar pessoas e depois se espalhar entre seres humanos como já ocorreu com o MERS-CoV e o SARS-CoV-2. Até o momento, não foi definido o reservatório silvestre do SARS-CoV-2.

Fonte: [Covid](https://www.gov.br/saude/pt-br/coronavirus/o-que-e-o-coronavirus)

### *Como se proteger*

Diante da emergência ocasionada pelo coronavírus SARS-CoV-2, o reconhecimento da pandemia pela OMS e a declaração de Emergência de Saúde Pública de Importância Nacional (ESPIN), o Ministério da Saúde tem estabelecido sistematicamente medidas para resposta e enfrentamento da covid-19.

Entre as medidas indicadas pelo MS, estão as não farmacológicas, como distanciamento social, etiqueta respiratória e de higienização das mãos, uso de máscaras, limpeza e desinfeção de ambientes, isolamento de casos suspeitos e confirmados e quarentena dos contatos dos casos de covid-19, conforme orientações médicas.

Ademais, o MS recomenda ainda a vacinação contra a covid-19 dos grupos prioritários conforme o Plano Nacional de Operacionalização da Vacinação. Estas medidas devem ser utilizadas de forma integrada, a fim de controlar a transmissão do SARSCoV-2, permitindo também a retomada gradual das atividades desenvolvidas pelos vários setores e o retorno seguro do convívio social.

### Valores ausentes

Nesta etapa, vamos verificar a quantidade de valores ausentes em nosso dataframe.

É possível notar a presença de dados ausentes, vamos analisar cada uma das variáveis que possuem uma grande quantidade de valores faltantes e definir o tratamento mais adequado.

Analisando os dados ausentes, é notável que:

**excess_mortality**, **total_boosters**, **total_boosters_per_hundred**, **weekly_icu_admissions,weekly_icu_admissions_per_million,weekly_hosp_admissions e weekly_hosp_admissions_per_million** - se tornam variáveis irrelevantes para a análise, uma vez que, cerca de 99% dos valores estão ausentes e por isso pode ser excluída. A variável tests_units, também será excluída pois não há relevância para a análise.

**icu_patients,icu_patients_per_million,hosp_patients e hosp_patients_per_million** - são variáveis referentes a um único dia de registro, possuem mais de 89% dos valores ausentes. Sendo assim, estes serão excluídos, afim de termos uma análise o mais próximo possível da realidade.

**people_fully_vaccinated,people_fully_vaccinated_per_hundred,new_vaccinations,people_vaccinated, people_vaccinated_per_hundred,total_vaccinations,total_vaccinations_per_hundred,new_vaccinations_smoothed_per_million e new_vaccinations_smoothed **- são variáveis referentes a vacinação, cujas campanhas começaram recentemente, explicando assim a quantidade de valores ausentes. Para análise, os valores faltantes serão preenchidos com 0.

```python
# Checar a porcentagem de valores ausentes
(df.isnull().sum()/ df.shape[0]).sort_values(ascending=False)
```

### Removendo Variáveis

Aqui iremos remover as variáveis irrelevantes para a análise, citadas acima.

Podemos ver que o DataFrame teve uma mudança considerável, de 62 colunas foi para 41 colunas, 21 delas possuíam a maioria de seus dados ausentes, isso seria irrelevante para a análise.

```python
# Criando cópia do DataFrame
df_clean = df.copy()

# Removendo as Variáveis
df_clean = df_clean.drop(['new_vaccinations_smoothed_per_million','new_vaccinations_smoothed','total_vaccinations_per_hundred','total_vaccinations','people_vaccinated_per_hundred','people_vaccinated','new_vaccinations','people_fully_vaccinated','people_fully_vaccinated_per_hundred','excess_mortality','total_boosters','total_boosters_per_hundred','weekly_icu_admissions','weekly_icu_admissions_per_million', 'weekly_hosp_admissions', 'weekly_hosp_admissions_per_million', 'tests_units', 'icu_patients', 'icu_patients_per_million', 'hosp_patients', 'hosp_patients_per_million'], axis=1)

# Checando corpo do DataFrame após a exclusão
print('Número de colunas antes da exclusão {}'. format(df.shape[1]))
print('Número de colunas após a exclusão {}'.format(df_clean.shape[1]))
```

```python
# Verificando se os Valores faltantes foram preenchidos
(df_clean.isnull().sum() / df.shape[0]).sort_values(ascending=False)
```

### (Q3) Visualização dos países com mais mortes (24/08/2021)

```python
# Localizando os dados sobre os Países
USA = df_clean.loc[df_clean.iso_code == 'USA']
IND = df_clean.loc[df_clean.iso_code == 'IND']
BRA = df_clean.loc[df_clean.iso_code == 'BRA']
FRA = df_clean.loc[df_clean.iso_code == 'FRA']
RUS = df_clean.loc[df_clean.iso_code == 'RUS']
MEX = df_clean.loc[df_clean.iso_code == 'MEX']

# Localizando os dados sobre o Mundo
WRL = df_clean.loc[df_clean.iso_code == 'OWID_WRL']

# Criando DataFrame com os Países selecionados
paises_grafico_casos = USA.append(IND).append(BRA).append(FRA).append(RUS)
paises_grafico_mortes = USA.append(BRA).append(MEX).append(IND)

# Criando DataFrame com os Países selecionados e o Mundo
paisesMundo_grafico_casos = WRL.append(USA).append(IND).append(BRA).append(FRA).append(RUS)
paisesMundo_grafico_mortes = WRL.append(USA).append(BRA).append(MEX).append(IND)

# Localizando Variáveis na data mais recente
total_casos = paises_grafico_casos.loc[paises_grafico_casos.date == '2021-08-24',['date', 'location', 'total_cases']].sort_values(by='total_cases', ascending=False)
total_mortes = paises_grafico_mortes.loc[paises_grafico_mortes.date == '2021-08-24',['date', 'location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)
total_casos_mundo = paisesMundo_grafico_casos.loc[paisesMundo_grafico_casos.date == '2021-08-24',['date', 'location', 'total_cases']].sort_values(by='total_cases', ascending=False)
total_mortes_mundo = paisesMundo_grafico_mortes.loc[paisesMundo_grafico_mortes.date == '2021-08-24',['date', 'location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)


# Plotando os gráficos de barras
fig, ax = plt.subplots(nrows=1 , ncols=4, figsize=(26,8))

# Definindo estilo dos Gráficos para o estilo do Seaborn
sns.set()

# Gráfico 1
sns.barplot(x='location', y=total_casos['total_cases'], data=paises_grafico_casos, ax=ax[0], palette='rocket')
ax[0].set_title('Número de Casos Por País')
ax[0].set_ylabel('Total de Casos')
ax[0].set_xlabel('País')

# Gráfico 2
sns.barplot(x='location', y=total_casos_mundo['total_cases'], data=paisesMundo_grafico_casos, ax=ax[1], palette='rocket')
ax[1].set_title('Número de Casos por País em Comparação ao Mundo')
ax[1].set_ylabel('Total de Casos')
ax[1].set_xlabel('Localização')

# Gráfico 3
sns.barplot(x='location', y=total_mortes['total_deaths'], data=paises_grafico_mortes, ax=ax[2], palette='rocket')
ax[2].set_title('Número de Mortes por País')
ax[2].set_ylabel('Total de Mortes')
ax[2].set_xlabel('País')

# Gráfico 4
sns.barplot(x='location', y=total_mortes_mundo['total_deaths'], data=paisesMundo_grafico_mortes, ax=ax[3], palette='rocket')
ax[3].set_title('Número de Mortes por Páis em Comparação ao Mundo')
ax[3].set_ylabel('Total de Mortes')
ax[3].set_xlabel('Localização')

plt.tight_layout();

# Salvar Imagem
fig.savefig('barrasCovidMundo.png')
```

```python
# Plotando os Gráficos de Linha
fig , ax = plt.subplots(nrows=4, ncols=1, figsize=(20,16))

# Gráfico 1
sns.lineplot(data=df_clean,x=df_clean['date'], y=df_clean['total_cases'] , hue=paises_grafico_casos['location'], palette='tab10' , ax=ax[0])
ax[0].set_title('Número de casos ao decorrer do Tempo')
ax[0].set_xlabel('Data')
ax[0].set_ylabel('Número de Casos')

# Gráfico 2
sns.lineplot(data=df_clean,x=df_clean['date'], y=df_clean['total_cases'], hue=paisesMundo_grafico_casos['location'], palette='tab10', ax=ax[1])
ax[1].set_title('Número de Casos ao Decorrer do Tempo em Relação ao Mundo')
ax[1].set_ylabel('Número de Casos')
ax[1].set_xlabel('Data')

# Gráfico 3
sns.lineplot(data=df_clean, x=df_clean['date'], y=df_clean['total_deaths'], hue=paises_grafico_mortes['location'], palette='tab10', ax=ax[2])
ax[2].set_title('Número de Mortes ao decorrer do Tempo')
ax[2].set_xlabel('Data')
ax[2].set_ylabel('Número de Mortes')

# Gráfico 4
sns.lineplot(data=df_clean, x=df_clean['date'], y=df_clean['total_deaths'], hue=paisesMundo_grafico_mortes['location'], palette='tab10', ax=ax[3])
ax[3].set_title('Número De Mortes ao Decorrer do Tempo em relação ao Mundo')
ax[3].set_ylabel('Número de mortes')
ax[3].set_xlabel('Data')

plt.tight_layout();

# Salvar Imagem
fig.savefig('linhasCovidMundo.png')
```
